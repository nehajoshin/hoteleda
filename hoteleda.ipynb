<b><h2>Q. Does higher lead time lead to booking cancelations?

````python
plt.figure(figsize = (12,8))
g = sns.FacetGrid(hotel_df_c, hue = 'is_canceled', height = 6, aspect = 2).map(sns.kdeplot, 'lead_time', shade = True).add_legend()
````

<b>When the lead time exceeds around 50 days, the cancellation rate among guests increases. To meet the needs of customers, it is crucial to provide accurate, real-time information and offer flexible booking options that allow for easy modifications. This will help to reduce the risk of cancellations and enhance customer satisfaction.

````python
cancel_lead_time  =hotel_df_c[hotel_df_c['is_canceled'] == 'canceled'].groupby('arrival_date_month')['lead_time'].median().reindex(ordered_month)
cancel_count = hotel_df_c[hotel_df_c['is_canceled'] == 'canceled'].groupby('arrival_date_month').size().reindex(ordered_month)

fig, ax = plt.subplots(figsize = (12,6))
ax = cancel_lead_time.plot(kind = 'line', marker = 'o')
ax.set_ylabel('Median Lead Time')
ax.legend(['Median Lead Time'], loc = 'upper left')
ax1 = ax.twinx()
ax1 = cancel_count.plot(kind = 'line', marker = 'v', color = 'green')
ax1.set_ylabel('Canceled bookings count')
ax1.legend(['Cancellations count'], loc = 'upper right')
plt.title('Median Lead Time and canceled bookings count per month')
````

<b>The data shows that as lead time increases, the number of cancellations in hotel bookings also tends to increase.
````python
def country_name(code):
    try:
        if (len(code)==2 and code in list_alpha_2):
            return pc.countries.get(alpha_2=code).name
        elif (len(code)==3 and code in list_alpha_3):
            return pc.countries.get(alpha_3=code).name
    except KeyError:
            return code
````
